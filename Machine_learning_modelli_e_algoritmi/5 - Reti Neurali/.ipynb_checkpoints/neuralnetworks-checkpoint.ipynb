## Reti Neurali - il Multilayer Perceptron

### Importiamo i moduli

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
!wget https://raw.githubusercontent.com/ProfAI/machine-learning-modelli-e-algoritmi/main/script/viz.py
```

```python
from viz import plot_decision_boundary
```

### Definiamo le costanti

```python
RANDOM_SEED = 0
```

### Generiamo i dati

```python
X, y = make_moons(n_samples=100, shuffle=True, noise=0.25, random_state=RANDOM_SEED)
plt.scatter(X[:,0], X[:,1], c=y)
```

### Processiamo i dati

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=RANDOM_SEED)
```

```python
mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
X_test = mms.transform(X_test)
```

### Perceptron

```python
from sklearn.linear_model import Perceptron
```

```python
model = Perceptron()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Regressione Logistica

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
# Visualizziamo il modello del Percettrone
plot_decision_boundary(model, (X_train, y_train), (X_test, y_test))
```

```python
# Visualizziamo il modello della Regressione Logistica
plot_decision_boundary(lr, (X_train, y_train), (X_test, y_test))
```

### Multilayer Perceptron

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100,100,100,), random_state=RANDOM_SEED) # 3 strati nascosti con 100 nodi
mlp.fit(X_train, y_train)
mlp.score(X_test, y_test)
```

```python
plot_decision_boundary(mlp, (X_train, y_train), (X_test, y_test))
```

Il modello commette ancora qualche errore ma approssima molto meglio la distribuzione dei dati

### Esploriamo la rete osservando i pesi

```python
len(mlp.coefs_) # numero strati nascosti + input 

# Per osservare i pesi del primo strato mlp.coefs_[0]
```

```python
len(mlp.intercepts_)
```

```python
# Calcoliamo quanti pesi e bias ci sono per ogni strato
weights = 0

for i in range(len(mlp.coefs_)):
  coefs =  mlp.coefs_[i].size
  bias = mlp.intercepts_[i].size
  tot = coefs+bias
  print(f"\tLayer {i} => pesi={coefs} bias={bias} totale={tot}")
  weights+=tot

print(f"\nPesi totali della rete={weights}")
```

### A cosa serve la funzione di attivazione?

Andiamo a toglierla definendola come funzione identità in una rete molto profonda con 10 strati nascosti con 100 nodi

```python
hiddel_layer_size = (100,)*10
mlp = MLPClassifier(hidden_layer_sizes=hiddel_layer_size, activation="identity", random_state=RANDOM_SEED)
mlp.fit(X_train, y_train)
```

```python
mlp.score(X_test, y_test)
```

Lo score è minore rispetto a prima (0.9) perchè rimuovendo la funzione di attivazione abbiamo rimosso qualsiasi tipo di possibilità di rilevare regressioni lineari all'interno di dati.

Adesso il modello è solo frutto di moltiplicazioni e somme, quindi rimaniamo all'interno di modelli lineari .

Quindi rimuovendo la funzione di attivazione perdiamo la possibilità di rilevare interazioni non lineari all'interno dei dati

```python
plot_decision_boundary(mlp, (X_train, y_train), (X_test, y_test))
```
