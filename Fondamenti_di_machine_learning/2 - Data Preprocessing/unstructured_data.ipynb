## Testo

```python
with open(BASE_URL+"chuck.txt", encoding="utf-8") as f:
    sentences = f.read().splitlines()

sentences
```

#### BOW - Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
data = cv.fit_transform(sentences)
data = data.toarray()

print(cv.vocabulary_)
print(data[0])
```

```python
bow = CountVectorizer(stop_words=["chuck","norris"])
data = bow.fit_transform(sentences)
data = data.toarray()

print(bow.vocabulary_)
print(data[0])
```

```python
bow = CountVectorizer(max_df=0.8, min_df=0.2)
data = bow.fit_transform(sentences)
data = data.toarray()

print(bow.vocabulary_)
print(data[0])
```

```python
bow = CountVectorizer(max_df=0.8, min_df=0.2)
data = bow.fit_transform(sentences)
data = data.toarray()

print(bow.vocabulary_)
print(data[0])
```

```python
bow = CountVectorizer(max_features=20)
data = bow.fit_transform(sentences)
data = data.toarray()

print(bow.vocabulary_)
print(data[0])
```

#### TF-IDF - Term frequency/Inverse document frequency

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
data = tfidf.fit_transform(sentences)
data = data.toarray()

print(tfidf.vocabulary_)
print(data[0])
```

```python
tfidf = TfidfVectorizer(stop_words=["chuck","norris"])
data = tfidf.fit_transform(sentences)
data = data.toarray()

print(tfidf.vocabulary_)
print(data[0])
```

```python
tfidf = TfidfVectorizer(max_df=0.8, min_df=0.2)
data = tfidf.fit_transform(sentences)
data = data.toarray()

print(tfidf.vocabulary_)
print(data[0])
```

```python
bow = TfidfVectorizer(max_features=20)
data = tfidf.fit_transform(sentences)
data = data.toarray()

print(tfidf.vocabulary_)
print(data[0])
```
